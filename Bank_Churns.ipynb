```python
# Visualization 1: Distribution of Credit Scores
plt.figure(figsize=(10, 3))
sns.histplot(df['CreditScore'], kde=True, bins=30, color='blue')
plt.title('Distribution of Credit Scores')
plt.xlabel('Credit Score')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualization 2: Gender Distribution
plt.figure(figsize=(5, 3))
sns.countplot(x='Gender', data=df, palette='pastel')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Visualization 3: Geography vs. Churn Rate
plt.figure(figsize=(8,3))
sns.barplot(x='Geography', y='Exited', data=df, palette='muted')
plt.title('Geography vs. Churn Rate')
plt.xlabel('Geography')
plt.ylabel('Average Churn Rate')
plt.show()
```

```python
# Visualization 4: Age Distribution by Churn Status
plt.figure(figsize=(12, 3))
sns.boxplot(x='Exited', y='Age', data=df, palette='coolwarm')
plt.title('Age Distribution by Churn Status')
plt.xlabel('Churn Status (Exited)')
plt.ylabel('Age')
plt.show()
```

                                              Conclusion
1. The distribution of credit scores suggests most customers have scores within a specific range.
2. The gender distribution may indicate an equal or skewed proportion of male and female customers.
3. Geography has a noticeable impact on churn rate.
4. Age distribution indicates older customers may be more likely to churn.
5. Correlation heatmap reveals relationships among numerical features.
